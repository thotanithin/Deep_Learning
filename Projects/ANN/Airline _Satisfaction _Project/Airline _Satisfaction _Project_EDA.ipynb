## Outiler treaments using IQR Method

```python
#Outiler treaments of  Flight Distance column
q3 = data['Flight Distance'].quantile(0.80)
q1 = data['Flight Distance'].quantile(0.20)
Iqr = q3-q1
upper = q3+(1.5*Iqr)
lower= q1-(1.5*Iqr)
print("Iqr = ",Iqr)
print("q3 = ",q3)
print("q1= ",q1)
print("upper_limit = ", upper)
print("lower_limit = ", lower)
```

```python
data = data[(data['Flight Distance']>=lower) & (data['Flight Distance']<= upper)]
```
